## visualization of one image of digit by converting columns into 28*28

```python
plt.figure(figsize=(25,25))
plt.subplot(1,10,1)
D=A.drop('label',axis=1).values
plt.title("One")
plt.imshow(D[0:1].reshape(28,28))


plt.subplot(1,10,2)
D=A.drop('label',axis=1)[1:2].values
plt.title("Two")
plt.imshow(D.reshape(28,28))

plt.subplot(1,10,3)
D=A.drop('label',axis=1)[2:3].values
plt.imshow(D.reshape(28,28))
plt.title("Tree")
plt.show()

plt.figure(figsize=(25,25))
plt.subplot(2,10,1)
D=A.drop('label',axis=1)[3:4].values
plt.title("Four")
plt.imshow(D.reshape(28,28))

plt.subplot(2,10,2)
D=A.drop('label',axis=1)[4:5].values
plt.title("Five")
plt.imshow(D.reshape(28,28))

plt.subplot(2,10,3)
D=A.drop('label',axis=1)[5:6].values
plt.imshow(D.reshape(28,28))
plt.title("Six")
plt.show()

plt.figure(figsize=(25,25))
plt.subplot(3,10,1)
D=A.drop('label',axis=1)[6:7].values
plt.title("Seven")
plt.imshow(D.reshape(28,28))

plt.subplot(3,10,2)
D=A.drop('label',axis=1)[7:8].values
plt.title("Eigth")
plt.imshow(D.reshape(28,28))

plt.subplot(3,10,3)
D=A.drop('label',axis=1)[8:9].values
plt.imshow(D.reshape(28,28))
plt.title("NINE")
plt.show()
```
